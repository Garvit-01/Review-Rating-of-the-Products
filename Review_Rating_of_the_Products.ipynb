```python
import nltk
import pandas as pd
import numpy as np
```

## Loading the Data

```python
df = pd.read_csv("Train_Data.csv")
df.head()
```

```python
df.shape
```

## Checking for null values 

```python
df.isnull().sum()
```

#### As there are 3 customer_reviews with null values so we should drop them 

```python
df = df.dropna(axis=0, subset=['customer_reviews'])
```

```python
## Shape after dropping 
df.shape
```

### Making function for defining customer rating 

```python
def rating_sentiment(n):
    if n>=4.5:
        return 5
    elif n>3.5:
        return 4
    elif n>2.5:
        return 3
    elif n>1.5:
        return 2
    elif n>0.5:
        return 1
    else:
        return 0
    
## Making new column called rating sentiment 
df['rating_sentiment'] = df['average_review_rating'].apply(rating_sentiment)
df
```

```python
## As new column have been added shape has been changed
df.shape
```

```python
# Customer reviews
X = df['customer_reviews']

#label
y = df['rating_sentiment']
```

```python
## Removing numbers and special characters 
X.replace("[^a-zA-Z]", " ", regex=True, inplace=True)
X.head
```

```python
X = X.str.lower()
X.head()
```

### CountVectorizer develops a vector of all the words in the string

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=3000)
X = cv.fit_transform(X).toarray()
```

```python
cv.get_feature_names()
```

```python
data = pd.DataFrame(X, columns=cv.get_feature_names())
data.shape
```

```python
X.shape
```

```python
y.shape
```

# TRANING ML MODELS

## Model 1 : RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier(n_estimators=200, criterion='gini', n_jobs=-1)
```

## Model 2 : Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB

model2 = MultinomialNB()
```

## Model 3: Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression()
```

## Model 4: AdaBoost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
model4 = AdaBoostClassifier()
```

## Model 5: KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model5 = KNeighborsClassifier()
```

## Calculating Accuracy using CrossValidationScore

```python
from sklearn.model_selection import cross_val_score

score1 = cross_val_score(model1, X, y, cv=5, verbose=1, n_jobs=-1)
score2 = cross_val_score(model2, X, y, cv=5, verbose=1, n_jobs=-1)
score3 = cross_val_score(model3, X, y, cv=5, verbose=1, n_jobs=-1)
score4 = cross_val_score(model4, X, y, cv=5, verbose=1, n_jobs=-1)
score5 = cross_val_score(model5, X, y, cv=5, verbose=1, n_jobs=-1)
```

```python
print("Random Forest Classifier (score1) = ", score1.mean())
print("Naive_bayes Multinomial (score2) = ", score2.mean())
print("LogisticRegression (score3) = ", score3.mean())
print("AdaBoostClassifier (score4) = ", score4.mean())
print("KNN (score5) = ", score5.mean())
```

### Random Forest Classifier model performs best with the accuracy of 80.074%
